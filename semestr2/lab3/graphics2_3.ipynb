```python
import numpy as np
import matplotlib.pyplot as plt

with open('numm2_3.txt', 'r') as f:
    lines = f.readlines()

n_values = []
results = []
for line in lines[1:-1]: 
    parts = line.strip().split('\t')
    n_values.append(int(parts[0]))
    results.append(float(parts[1]))

exact_value = float(lines[-1].split()[-1])
print(exact_value)
errors = [abs(result - exact_value) for result in results]

plt.figure(figsize=(10, 6))
plt.loglog(n_values, errors, 'bo-', label='')
plt.xlabel('  (n)', fontsize=12)
plt.ylabel(' ', fontsize=12)
plt.title('    ', fontsize=14)
plt.grid(True, which="both", ls="-")
plt.legend(fontsize=12)

n_theory = np.array(n_values)
plt.loglog(n_theory, 10/n_theory, 'r--', label=' O(1/n)')
plt.legend()

plt.show()
```
