```python
import matplotlib.pyplot as plt
import numpy as np

def plot_error_vs_step(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()[1:]  
    h_values = []
    errors = []
    for line in lines:
        if line.strip():
            parts = line.split()
            h_values.append(int(parts[0])) 
            errors.append(float(parts[1]))  
    
    actual_steps = [1.0/h for h in h_values] 

    plt.figure(figsize=(12, 6))
    plt.gca().invert_xaxis()  
    plt.loglog(actual_steps, errors, 'ro-', label='  (RK3)')

    C = errors[0] / (actual_steps[0]**3)
    theoretical = [C * h**3 for h in actual_steps]
    plt.loglog(actual_steps, theoretical, 'b--', label=' O(h)')
    
    plt.xlabel('')
    plt.ylabel(' ')
    plt.title('    ( )')
    plt.grid(True, which="both", ls="-")
    plt.legend()
    
   
    
    plt.tight_layout()
    plt.show()

plot_error_vs_step(r"C:\Users\yatma\source\repos\numm2_5\numm2_5\errors.txt")
```
